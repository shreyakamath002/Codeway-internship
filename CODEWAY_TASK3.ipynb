```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#Importing the dataset
df=pd.read_csv("D:\Internship\Churn_Modelling.csv")
df
```

```python
df.head(10)
```

```python
df.describe()
```

```python
##Data Preprocessing
```

```python
#The target variable representing customer churn is "Excited"
```

```python
x=df.drop(['Exited','RowNumber','CustomerId','Surname'],axis=1)
y=df['Exited']
```

```python
#Converting the categorical variables to dummy variables
```

```python
x=pd.get_dummies(x,columns=['Geography','Gender'],drop_first=True)
```

```python
#Split the dataset into testing and training sets
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Interpretation:These subsets are typically used for training and evaluating machine learning models.The training set is used to train the model, while the testing set is used to evaluate its performance on unseen data. Splitting the data into training and testing sets helps assess how well the model generalizes to new, unseen data and prevents overfitting.
```

```python
##Feature scaling
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
##Building and training Logistic Regression Model 
```

```python
logistic_classifier=LogisticRegression(random_state=42)
logistic_classifier.fit(x_train,y_train)
```

```python
##Building and training Random Forest Classifier Model 
```

```python
random_forest_classifier=RandomForestClassifier(random_state=42)
random_forest_classifier.fit(x_train,y_train)
```

```python
##Building and training Gradient Boosting Classifier Model
```

```python
gradient_boosting_classifier=GradientBoostingClassifier(random_state=42)
gradient_boosting_classifier.fit(x_train,y_train)
```

```python
##Evaluating Logistic Regression Model
logistic_y_pred=logistic_classifier.predict(x_test)
logistic_accuracy=accuracy_score(y_test,logistic_y_pred)
logistic_conf_matrix=confusion_matrix(y_test,logistic_y_pred)
logistic_class_report=classification_report(y_test,logistic_y_pred)
```

```python
##Evaluating Random Forest Classifier Model
random_forest_y_pred=random_forest_classifier.predict(x_test)
random_forest_accuracy=accuracy_score(y_test,random_forest_y_pred)
random_forest_conf_matrix=confusion_matrix(y_test,random_forest_y_pred)
random_forest_class_report=classification_report(y_test,random_forest_y_pred)
```

```python
##Evaluating Gradient Boosting Classifier Model
gradient_boosting_y_pred=gradient_boosting_classifier.predict(x_test)
gradient_boosting_accuracy=accuracy_score(y_test,gradient_boosting_y_pred)
gradient_boosting_conf_matrix=confusion_matrix(y_test,gradient_boosting_y_pred)
gradient_boosting_class_report=classification_report(y_test,gradient_boosting_y_pred)
```

```python
##Model Evaluation Metrics for Logistic Regression
```

```python
print("\nLogistic Regression Model Evaluation Metrics:")
print(f"Accuracy:{logistic_accuracy:.3f}")
```

```python
sns.countplot(x='Exited',hue='Gender',data=df)
plt.title('Churn count by Gender')
plt.xlabel('Churn(0:No Churn,1: Churn)')
plt.ylabel('Count')
plt.show()
```

##### The count plot visualizes the distribution of churn (Exited) for each gender category.The 'Exited' variable is plotted on the x-axis, where 0 typically represents no churn (not exited), and 1 represents churn (exited).The plot provides insights into the churn behavior of customers based on gender and can help in understanding if there are any gender-related patterns in customer churn.
